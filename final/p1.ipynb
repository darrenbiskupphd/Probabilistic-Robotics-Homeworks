```python
import numpy as np
from sympy import *
np.set_printoptions(precision=4, suppress=True)
```

```python
px, py = symbols("p_x p_y")
h = sqrt((4 - px)**2 + (5 - py)**2)
Hx = h.diff(Matrix([px,py]))
simplify(Hx)
```

```python
x_bar = np.array([1,1])
Sigmax_bar = np.array([[0.2**2 , 0.1*0.2**2],
                       [0.1*0.2**2, 0.2**2]])
Sigmaz = 0.1**2

Hx = np.array(Hx.subs({px: 1, py: 1})).astype(np.float64).flatten()
K = Sigmax_bar @ Hx.T * 1/(Hx @ Sigmax_bar @ Hx.T + Sigmaz)
K
```

```python
x = x_bar + K * (4.8 - 5)
x
```

```python
### posterior sigmax
(np.eye(2) - K @ Hx) @ Sigmax_bar
```
