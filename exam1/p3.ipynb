### declare system parameters

```python
import numpy as np
from sympy import *
np.set_printoptions(precision=5, suppress=True)
```

```python
Kt , J, b , R , L = symbols('K_t , J, b , R , L')
rho, e = symbols('rho, e')
dt = symbols('dt')

A = Matrix([[1, dt, 0],
            [0, 1 - dt*b/J, dt*Kt/J],
            [0, -dt* Kt/L, 1-dt*R/L]])

B = Matrix([[0,0],
              [-dt/J, 0],
              [0, dt/L]])
C = Matrix([[rho/(2*pi), 0, 0]])
```

```python
sigma_T, sigma_V = symbols('sigma_T, sigma_V')
S_u = Matrix([[sigma_T**2, 0],
              [0, sigma_V**2]])
Sigma_xbar = B*S_u*B.T
Sigma_xbar
```

```python
sigma_e = symbols('sigma_e')
S_z = Matrix([sigma_e**2])

K = Sigma_xbar*C.T*(C*Sigma_xbar*C.T + S_z).inv()
K
```
