```python
from sympy import *
import numpy as np
np.set_printoptions(precision=6, suppress=True)
```

```python
x = Matrix([10,13,30])
S = Matrix.diag([0.3**2, 0.3**2, 10**2])
```

```python
L = 3
lam = 9
kappa = 0
beta = 3

x0 = x

x1 = (x + ((L+lam)*S).pow(0.5)[0,:].T)
x2 = (x + ((L+lam)*S).pow(0.5)[1,:].T)
x3 = (x + ((L+lam)*S).pow(0.5)[2,:].T)

x4 = (x - ((L+lam)*S).pow(0.5)[0,:].T)
x5 = (x - ((L+lam)*S).pow(0.5)[1,:].T)
x6 = (x - ((L+lam)*S).pow(0.5)[2,:].T)
```

```python
R = Matrix([[cos(rad(x[2])), -sin(rad(x[2])), 0],
            [sin(rad(x[2])), cos(rad(x[2])), 0],
            [0,0,1]])

gamma0 = x0 + R*Matrix([0.3, 0.2, 0])

wm = 0.75
wc = 0.75

x_bar_0 = wm * gamma0
x_bar_0.evalf()
```

```python
S_bar = wc*(gamma0 - x_bar_0)*(gamma0 - x_bar_0).T
S_bar.evalf()
```
